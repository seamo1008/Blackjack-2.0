```python
class PLAYER:
    def __init__(self):
        self.deck = DECK()
        player_name = None
        self.name = player_name
        self.wallet = 100 #Initial wallet balance
        self.bet_amount = 0 #Initial bet amount
        self.hand = []
        self.hand_2 = []
        self.dealer_hand = []
        self.hand_value = self.calculate_player_hand_value()
        self.dealer_hand_value = self.calculate_dealer_hand_value()
        self.hand_size = 2
        self.bet_confirmed = False
        self.bet_deducted = False
        self.dealer_hit = False
        self.dealer_stay = False
        self.dealer_blackjack = False
        self.dealer_bust = False
        self.dealer_wins = False
        self.player_stay = False
        self.player_bust = False
        self.player_blackjack = False
        self.player_double = False
        self.player_wins = False
        self.push = False
        self.hand_calculation_done = False

        self.five_table = pygame.image.load('Graphics/5_table.png').convert_alpha()
        self.ten_table = pygame.image.load('Graphics/10_table.png').convert_alpha()
        self.fifteen_table = pygame.image.load('Graphics/15_table.png').convert_alpha()
        self.twenty_table = pygame.image.load('Graphics/20_table.png').convert_alpha()
        self.twenty_five_table = pygame.image.load('Graphics/25_table.png').convert_alpha()
        self.thirty_table = pygame.image.load('Graphics/30_table.png').convert_alpha()
        self.thirty_five_table = pygame.image.load('Graphics/35_table.png').convert_alpha()
        self.forty_table = pygame.image.load('Graphics/40_table.png').convert_alpha()
        self.forty_five_table = pygame.image.load('Graphics/45_table.png').convert_alpha()
        self.fifty_table = pygame.image.load('Graphics/50_table.png').convert_alpha()
        self.null_table_text = (".")
        self.null_table_surface = small_font.render(self.null_table_text,True,(255,255,255))
        self.null_table_rect = self.null_table_surface.get_rect(center = (width/2,height-250))
        self.five_table_rect = self.five_table.get_rect(center = ((width / 2), height-250))
        self.ten_table_rect = self.ten_table.get_rect(center = ((width / 2), height-250))
        self.fifteen_table_rect = self.fifteen_table.get_rect(center = ((width / 2), height-250))
        self.twenty_table_rect = self.twenty_table.get_rect(center = ((width / 2), height-250))
        self.twenty_five_table_rect = self.twenty_five_table.get_rect(center = ((width / 2), height-250))
        self.thirty_table_rect = self.thirty_table.get_rect(center = ((width / 2), height-250))
        self.thirty_five_table_rect = self.thirty_five_table.get_rect(center = ((width / 2), height-250))
        self.forty_table_rect = self.forty_table.get_rect(center = ((width / 2), height-250))
        self.forty_five_table_rect = self.forty_five_table.get_rect(center = ((width / 2), height-250))
        self.fifty_table_rect = self.fifty_table.get_rect(center = ((width / 2), height-250))

        



    def enter_name(self, name):
        self.name = name

    def bet_five(self):
        self.bet_amount += 5
        if self.bet_amount > self.wallet:
            print("No No No. Da. Da li imas pare? Balance too low for this bet. No")
            self.bet_amount -= 5

    def bet_ten(self):
        self.bet_amount += 10
        if self.bet_amount > self.wallet:
            print("No No No. Da. Da li imas pare? Balance too low for this bet. No")
            self.bet_amount -= 10

    def bet_twenty(self):
        self.bet_amount += 20
        if self.bet_amount > self.wallet:
            print("No No No. Da. Da li imas pare? Balance too low for this bet. No")
            self.bet_amount -= 20

    def clear_bet(self):
        self.bet_amount = 0

    def place_bet(self):
        if not self.bet_confirmed:
            pass
        else:
            self.wallet -= self.bet_amount
            self.bet_confirmed = False
            self.bet_deducted = True

    def draw_table_bet(self):            
        table_bet = int(self.bet_amount)
        if table_bet == 5:
            image,rect = self.five_table,self.five_table_rect
        elif table_bet == 10:
            image,rect = self.ten_table,self.ten_table_rect
        elif table_bet == 15:
            image,rect = self.fifteen_table,self.fifteen_table_rect
        elif table_bet == 20:
            image,rect = self.twenty_table,self.twenty_table_rect
        elif table_bet == 25:
            image,rect = self.twenty_five_table,self.twenty_five_table_rect
        elif table_bet == 30:
            image,rect = self.thirty_table,self.thirty_table_rect
        elif table_bet == 35:
            image,rect = self.thirty_five_table,self.thirty_five_table_rect
        elif table_bet == 40:
            image,rect = self.forty_table,self.forty_table_rect
        elif table_bet == 45:
            image,rect = self.forty_five_table,self.forty_five_table_rect
        elif table_bet >= 50:
            image,rect = self.fifty_table,self.fifty_table_rect
        else:
            image,rect = self.null_table_surface,self.null_table_rect
        return image,rect

    def calculate_player_hand_value(self):
        total_value = 0
        num_aces = 0

        for i, card in enumerate(self.hand):
            ace_check = card[0]
            rank = card[1]
            if ace_check == "Ace of Spades" or ace_check == "Ace of Hearts" or ace_check == "Ace of Clubs" or ace_check == "Ace of Diamonds":
                num_aces += 1
            else:    
                total_value += rank

        for _ in range(num_aces):
            if total_value + 11 <= 21:
                total_value += 11
            else:
                total_value += 1
        
        self.hand_value = total_value

        return total_value
    
    def calculate_dealer_hand_value(self):
        total_value = 0
        num_aces = 0

        for i, card in enumerate(self.dealer_hand):
            ace_check = card[0]
            rank = card[1]
            if ace_check == "Ace of Spades" or ace_check == "Ace of Hearts" or ace_check == "Ace of Clubs" or ace_check == "Ace of Diamonds":
                num_aces += 1
            else:    
                total_value += rank

        for _ in range(num_aces):
            if total_value + 11 <= 21:
                total_value += 11
            else:
                total_value += 1

        self.dealer_hand_value = total_value
        
        return total_value

    def draw_player_hand_value(self):
        hand_value_text = self.calculate_player_hand_value()
        hand_value_surface = font.render(str(hand_value_text),True,(255,255,255))
        hand_value_rect = hand_value_surface.get_rect(center = ((width/2)+75,height-170))
        backing_colour = pygame.Surface.get_at(screen,(hand_value_rect.topright))
        screen.fill((backing_colour),hand_value_rect)
        screen.blit(hand_value_surface,hand_value_rect)

    def draw_dealer_hand_value(self):
        dealer_hand_value = self.calculate_dealer_hand_value()
        dealer_hand_value_surface = font.render(str(dealer_hand_value),True,(255,255,255))
        dealer_hand_value_rect = dealer_hand_value_surface.get_rect(center = ((width/2)+75,97))
        backing_colour = pygame.Surface.get_at(screen,(dealer_hand_value_rect.bottomright))
        screen.fill((backing_colour),dealer_hand_value_rect)
        screen.blit(dealer_hand_value_surface,dealer_hand_value_rect)

    def draw_wallet(self):
        wallet_text = self.wallet
        wallet_surface = small_font.render(str(wallet_text),True,(102,205,0))
        wallet_rect = wallet_surface.get_rect(center = (100,40))
        chip_stack_rect = chip_stack.get_rect(midright = (wallet_rect.left,wallet_rect.centery))
        screen.blit(wallet_surface,wallet_rect)
        screen.blit(chip_stack,chip_stack_rect)

    def draw_player_name(self):
        name_text = self.name
        name_surface = font.render(name_text,True,(255,255,255))
        name_rect = name_surface.get_rect(center = (width / 2,height - 170))
        screen.blit(name_surface,name_rect)

    def draw_dealer_name(self):
        dealer_text = "Dealer"
        dealer_surface = font.render(dealer_text,True,(255,255,255))
        dealer_rect = dealer_surface.get_rect(center = (width/2,95))
        screen.blit(dealer_surface,dealer_rect)

    def draw_bet_amount(self):
        bet_amount_text = self.bet_amount
        bet_amount_surface = small_font.render(str(bet_amount_text),True,(102,205,0))
        bet_amount_rect = bet_amount_surface.get_rect(center = (100,80))
        bet_symbol = small_font.render("BET: ",True,(102,205,0))
        bet_symbol_rect = bet_symbol.get_rect(midright = (bet_amount_rect.left,bet_amount_rect.centery))
        screen.blit(bet_amount_surface,bet_amount_rect)
        screen.blit(bet_symbol,bet_symbol_rect)

    def draw_dealer_hand(self):
        if len(self.dealer_hand) == 0:
            pass
        elif len(self.dealer_hand) == 1:
            dealer_card_one = self.dealer_hand[0]
            dealer_card_one_surface = self.deck.card_images.get(dealer_card_one[0])
            dealer_card_one_rect = dealer_card_one_surface.get_rect(center = ((width/2) - 25,165))
            if dealer_card_one_surface is not None:
                screen.blit(dealer_card_one_surface,dealer_card_one_rect)
        elif len(self.dealer_hand) == 2:
            dealer_card_one = self.dealer_hand[0]
            dealer_card_one_surface = self.deck.card_images.get(dealer_card_one[0])
            dealer_card_one_rect = dealer_card_one_surface.get_rect(center = ((width/2) - 25,165))
            dealer_card_two = self.dealer_hand[1]
            dealer_card_two_surface = self.deck.card_images.get(dealer_card_two[0])
            dealer_card_two_rect = dealer_card_two_surface.get_rect(center = ((width/2),192))
            if dealer_card_two_surface is not None:
                screen.blit(dealer_card_one_surface,dealer_card_one_rect)
                screen.blit(dealer_card_two_surface,dealer_card_two_rect)
        elif len(self.dealer_hand) == 3:
            dealer_card_one = self.dealer_hand[0]
            dealer_card_one_surface = self.deck.card_images.get(dealer_card_one[0])
            dealer_card_one_rect = dealer_card_one_surface.get_rect(center = ((width/2) - 25,165))
            dealer_card_two = self.dealer_hand[1]
            dealer_card_two_surface = self.deck.card_images.get(dealer_card_two[0])
            dealer_card_two_rect = dealer_card_two_surface.get_rect(center = ((width/2),192))
            dealer_card_three = self.dealer_hand[2]
            dealer_card_three_surface = self.deck.card_images.get(dealer_card_three[0])
            dealer_card_three_rect = dealer_card_three_surface.get_rect(center = ((width/2)+25,219))
            if dealer_card_three_surface is not None:
                screen.blit(dealer_card_one_surface,dealer_card_one_rect)
                screen.blit(dealer_card_two_surface,dealer_card_two_rect)
                screen.blit(dealer_card_three_surface,dealer_card_three_rect)
        elif len(self.dealer_hand) == 4:
            dealer_card_one = self.dealer_hand[0]
            dealer_card_one_surface = self.deck.card_images.get(dealer_card_one[0])
            dealer_card_one_rect = dealer_card_one_surface.get_rect(center = ((width/2) - 25,165))
            dealer_card_two = self.dealer_hand[1]
            dealer_card_two_surface = self.deck.card_images.get(dealer_card_two[0])
            dealer_card_two_rect = dealer_card_two_surface.get_rect(center = ((width/2),192))
            dealer_card_three = self.dealer_hand[2]
            dealer_card_three_surface = self.deck.card_images.get(dealer_card_three[0])
            dealer_card_three_rect = dealer_card_three_surface.get_rect(center = ((width/2)+25,219))
            dealer_card_four = self.dealer_hand[3]
            dealer_card_four_surface = self.deck.card_images.get(dealer_card_four[0])
            dealer_card_four_rect = dealer_card_four_surface.get_rect(midleft = dealer_card_one_rect.midright)
            if dealer_card_four_surface is not None:
                screen.blit(dealer_card_one_surface,dealer_card_one_rect)
                screen.blit(dealer_card_two_surface,dealer_card_two_rect)
                screen.blit(dealer_card_three_surface,dealer_card_three_rect)
                screen.blit(dealer_card_four_surface,dealer_card_four_rect)
        elif len(self.dealer_hand) == 5:
            dealer_card_one = self.dealer_hand[0]
            dealer_card_one_surface = self.deck.card_images.get(dealer_card_one[0])
            dealer_card_one_rect = dealer_card_one_surface.get_rect(center = ((width/2) - 25,165))
            dealer_card_two = self.dealer_hand[1]
            dealer_card_two_surface = self.deck.card_images.get(dealer_card_two[0])
            dealer_card_two_rect = dealer_card_two_surface.get_rect(center = ((width/2),192))
            dealer_card_three = self.dealer_hand[2]
            dealer_card_three_surface = self.deck.card_images.get(dealer_card_three[0])
            dealer_card_three_rect = dealer_card_three_surface.get_rect(center = ((width/2)+25,219))
            dealer_card_four = self.dealer_hand[3]
            dealer_card_four_surface = self.deck.card_images.get(dealer_card_four[0])
            dealer_card_four_rect = dealer_card_four_surface.get_rect(midleft = dealer_card_one_rect.midright)
            dealer_card_five = self.dealer_hand[4]
            dealer_card_five_surface = self.deck.card_images.get(dealer_card_five[0])
            dealer_card_five_rect = dealer_card_five_surface.get_rect(midleft = dealer_card_two_rect.midright)
            if dealer_card_five_surface is not None:
                screen.blit(dealer_card_one_surface,dealer_card_one_rect)
                screen.blit(dealer_card_two_surface,dealer_card_two_rect)
                screen.blit(dealer_card_three_surface,dealer_card_three_rect)
                screen.blit(dealer_card_four_surface,dealer_card_four_rect)
                screen.blit(dealer_card_five_surface,dealer_card_five_rect)
        elif len(self.dealer_hand) == 6:
            dealer_card_one = self.dealer_hand[0]
            dealer_card_one_surface = self.deck.card_images.get(dealer_card_one[0])
            dealer_card_one_rect = dealer_card_one_surface.get_rect(center = ((width/2) - 25,165))
            dealer_card_two = self.dealer_hand[1]
            dealer_card_two_surface = self.deck.card_images.get(dealer_card_two[0])
            dealer_card_two_rect = dealer_card_two_surface.get_rect(center = ((width/2),192))
            dealer_card_three = self.dealer_hand[2]
            dealer_card_three_surface = self.deck.card_images.get(dealer_card_three[0])
            dealer_card_three_rect = dealer_card_three_surface.get_rect(center = ((width/2)+25,219))
            dealer_card_four = self.dealer_hand[3]
            dealer_card_four_surface = self.deck.card_images.get(dealer_card_four[0])
            dealer_card_four_rect = dealer_card_four_surface.get_rect(midleft = dealer_card_one_rect.midright)
            dealer_card_five = self.dealer_hand[4]
            dealer_card_five_surface = self.deck.card_images.get(dealer_card_five[0])
            dealer_card_five_rect = dealer_card_five_surface.get_rect(midleft = dealer_card_two_rect.midright)
            dealer_card_six = self.dealer_hand[5]
            dealer_card_six_surface = self.deck.card_images.get(dealer_card_six[0])
            dealer_card_six_rect = dealer_card_six_surface.get_rect(midleft = dealer_card_three_rect.midright)
            if dealer_card_six_surface is not None:
                screen.blit(dealer_card_one_surface,dealer_card_one_rect)
                screen.blit(dealer_card_two_surface,dealer_card_two_rect)
                screen.blit(dealer_card_three_surface,dealer_card_three_rect)
                screen.blit(dealer_card_four_surface,dealer_card_four_rect)
                screen.blit(dealer_card_five_surface,dealer_card_five_rect)
                screen.blit(dealer_card_six_surface,dealer_card_six_rect)
        elif len(self.dealer_hand) == 7:
            dealer_card_one = self.dealer_hand[0]
            dealer_card_one_surface = self.deck.card_images.get(dealer_card_one[0])
            dealer_card_one_rect = dealer_card_one_surface.get_rect(center = ((width/2) - 25,165))
            dealer_card_two = self.dealer_hand[1]
            dealer_card_two_surface = self.deck.card_images.get(dealer_card_two[0])
            dealer_card_two_rect = dealer_card_two_surface.get_rect(center = ((width/2),192))
            dealer_card_three = self.dealer_hand[2]
            dealer_card_three_surface = self.deck.card_images.get(dealer_card_three[0])
            dealer_card_three_rect = dealer_card_three_surface.get_rect(center = ((width/2)+25,219))
            dealer_card_four = self.dealer_hand[3]
            dealer_card_four_surface = self.deck.card_images.get(dealer_card_four[0])
            dealer_card_four_rect = dealer_card_four_surface.get_rect(midleft = dealer_card_one_rect.midright)
            dealer_card_five = self.dealer_hand[4]
            dealer_card_five_surface = self.deck.card_images.get(dealer_card_five[0])
            dealer_card_five_rect = dealer_card_five_surface.get_rect(midleft = dealer_card_two_rect.midright)
            dealer_card_six = self.dealer_hand[5]
            dealer_card_six_surface = self.deck.card_images.get(dealer_card_six[0])
            dealer_card_six_rect = dealer_card_six_surface.get_rect(midleft = dealer_card_three_rect.midright)
            dealer_card_seven = self.dealer_hand[6]
            dealer_card_seven_surface = self.deck.card_images.get(dealer_card_seven[0])
            dealer_card_seven_rect = dealer_card_seven_surface.get_rect(midleft = dealer_card_four_rect.midright)
            if dealer_card_six_surface is not None:
                screen.blit(dealer_card_one_surface,dealer_card_one_rect)
                screen.blit(dealer_card_two_surface,dealer_card_two_rect)
                screen.blit(dealer_card_three_surface,dealer_card_three_rect)
                screen.blit(dealer_card_four_surface,dealer_card_four_rect)
                screen.blit(dealer_card_five_surface,dealer_card_five_rect)
                screen.blit(dealer_card_six_surface,dealer_card_six_rect)
                screen.blit(dealer_card_seven_surface,dealer_card_seven_rect)

    def draw_player_hand(self):
        if len(self.hand) == 0:
            pass
        elif len(self.hand) == 2:
            player_card_one = self.hand[0]
            player_card_one_surface = self.deck.card_images.get(player_card_one[0])
            player_card_one_rect = player_card_one_surface.get_rect(center = ((width/2)-40,height-300))
            player_card_two = self.hand[1]
            player_card_two_surface = self.deck.card_images.get(player_card_two[0])
            player_card_two_rect = player_card_two_surface.get_rect(center = ((width/2)-15,height-273))
            if player_card_one_surface is not None:
                screen.blit(player_card_one_surface,player_card_one_rect)
                screen.blit(player_card_two_surface,player_card_two_rect)
        elif len(self.hand) == 3:
            player_card_one = self.hand[0]
            player_card_one_surface = self.deck.card_images.get(player_card_one[0])
            player_card_one_rect = player_card_one_surface.get_rect(center = ((width/2)-40,height-300))
            player_card_two = self.hand[1]
            player_card_two_surface = self.deck.card_images.get(player_card_two[0])
            player_card_two_rect = player_card_two_surface.get_rect(center = ((width/2)-15,height-273))
            player_card_three = self.hand[2]
            player_card_three_surface = self.deck.card_images.get(player_card_three[0])
            player_card_three_rect = player_card_three_surface.get_rect(center = ((width/2)+10,height-246))
            if player_card_three_surface is not None:
                screen.blit(player_card_one_surface,player_card_one_rect)
                screen.blit(player_card_two_surface,player_card_two_rect)
                screen.blit(player_card_three_surface,player_card_three_rect)
        elif len(self.hand) == 4:
            player_card_one = self.hand[0]
            player_card_one_surface = self.deck.card_images.get(player_card_one[0])
            player_card_one_rect = player_card_one_surface.get_rect(center = ((width/2)-40,height-300))
            player_card_two = self.hand[1]
            player_card_two_surface = self.deck.card_images.get(player_card_two[0])
            player_card_two_rect = player_card_two_surface.get_rect(center = ((width/2)-15,height-273))
            player_card_three = self.hand[2]
            player_card_three_surface = self.deck.card_images.get(player_card_three[0])
            player_card_three_rect = player_card_three_surface.get_rect(center = ((width/2)+10,height-246))
            player_card_four = self.hand[3]
            player_card_four_surface = self.deck.card_images.get(player_card_four[0])
            player_card_four_rect = player_card_four_surface.get_rect(midleft = player_card_one_rect.midright)
            if player_card_four_surface is not None:
                screen.blit(player_card_one_surface,player_card_one_rect)
                screen.blit(player_card_two_surface,player_card_two_rect)
                screen.blit(player_card_three_surface,player_card_three_rect)
                screen.blit(player_card_four_surface,player_card_four_rect)
        elif len(self.hand) == 5:
            player_card_one = self.hand[0]
            player_card_one_surface = self.deck.card_images.get(player_card_one[0])
            player_card_one_rect = player_card_one_surface.get_rect(center = ((width/2)-40,height-300))
            player_card_two = self.hand[1]
            player_card_two_surface = self.deck.card_images.get(player_card_two[0])
            player_card_two_rect = player_card_two_surface.get_rect(center = ((width/2)-15,height-273))
            player_card_three = self.hand[2]
            player_card_three_surface = self.deck.card_images.get(player_card_three[0])
            player_card_three_rect = player_card_three_surface.get_rect(center = ((width/2)+10,height-246))
            player_card_four = self.hand[3]
            player_card_four_surface = self.deck.card_images.get(player_card_four[0])
            player_card_four_rect = player_card_four_surface.get_rect(midleft = player_card_one_rect.midright)
            player_card_five = self.hand[4]
            player_card_five_surface = self.deck.card_images.get(player_card_five[0])
            player_card_five_rect = player_card_five_surface.get_rect(midleft = player_card_two_rect.midright)
            if player_card_five_surface is not None:
                screen.blit(player_card_one_surface,player_card_one_rect)
                screen.blit(player_card_two_surface,player_card_two_rect)
                screen.blit(player_card_three_surface,player_card_three_rect)
                screen.blit(player_card_four_surface,player_card_four_rect)
                screen.blit(player_card_five_surface,player_card_five_rect)
        elif len(self.hand) == 6:
            player_card_one = self.hand[0]
            player_card_one_surface = self.deck.card_images.get(player_card_one[0])
            player_card_one_rect = player_card_one_surface.get_rect(center = ((width/2)-40,height-300))
            player_card_two = self.hand[1]
            player_card_two_surface = self.deck.card_images.get(player_card_two[0])
            player_card_two_rect = player_card_two_surface.get_rect(center = ((width/2)-15,height-273))
            player_card_three = self.hand[2]
            player_card_three_surface = self.deck.card_images.get(player_card_three[0])
            player_card_three_rect = player_card_three_surface.get_rect(center = ((width/2)+10,height-246))
            player_card_four = self.hand[3]
            player_card_four_surface = self.deck.card_images.get(player_card_four[0])
            player_card_four_rect = player_card_four_surface.get_rect(midleft = player_card_one_rect.midright)
            player_card_five = self.hand[4]
            player_card_five_surface = self.deck.card_images.get(player_card_five[0])
            player_card_five_rect = player_card_five_surface.get_rect(midleft = player_card_two_rect.midright)
            player_card_six = self.hand[5]
            player_card_six_surface = self.deck.card_images.get(player_card_six[0])
            player_card_six_rect = player_card_six_surface.get_rect(midleft = player_card_three_rect.midright)
            if player_card_six_surface is not None:
                screen.blit(player_card_one_surface,player_card_one_rect)
                screen.blit(player_card_two_surface,player_card_two_rect)
                screen.blit(player_card_three_surface,player_card_three_rect)
                screen.blit(player_card_four_surface,player_card_four_rect)
                screen.blit(player_card_five_surface,player_card_five_rect)
                screen.blit(player_card_six_surface,player_card_six_rect)
        elif len(self.hand) == 7:
            player_card_one = self.hand[0]
            player_card_one_surface = self.deck.card_images.get(player_card_one[0])
            player_card_one_rect = player_card_one_surface.get_rect(center = ((width/2)-40,height-300))
            player_card_two = self.hand[1]
            player_card_two_surface = self.deck.card_images.get(player_card_two[0])
            player_card_two_rect = player_card_two_surface.get_rect(center = ((width/2)-15,height-273))
            player_card_three = self.hand[2]
            player_card_three_surface = self.deck.card_images.get(player_card_three[0])
            player_card_three_rect = player_card_three_surface.get_rect(center = ((width/2)+10,height-246))
            player_card_four = self.hand[3]
            player_card_four_surface = self.deck.card_images.get(player_card_four[0])
            player_card_four_rect = player_card_four_surface.get_rect(midleft = player_card_one_rect.midright)
            player_card_five = self.hand[4]
            player_card_five_surface = self.deck.card_images.get(player_card_five[0])
            player_card_five_rect = player_card_five_surface.get_rect(midleft = player_card_two_rect.midright)
            player_card_six = self.hand[5]
            player_card_six_surface = self.deck.card_images.get(player_card_six[0])
            player_card_six_rect = player_card_six_surface.get_rect(midleft = player_card_three_rect.midright)
            player_card_seven = self.hand[6]
            player_card_seven_surface = self.deck.card_images.get(player_card_seven[0])
            player_card_seven_rect = player_card_seven_surface.get_rect(midleft = player_card_four_rect.midright)
            if player_card_six_surface is not None:
                screen.blit(player_card_one_surface,player_card_one_rect)
                screen.blit(player_card_two_surface,player_card_two_rect)
                screen.blit(player_card_three_surface,player_card_three_rect)
                screen.blit(player_card_four_surface,player_card_four_rect)
                screen.blit(player_card_five_surface,player_card_five_rect)
                screen.blit(player_card_six_surface,player_card_six_rect)
                screen.blit(player_card_seven_surface,player_card_seven_rect)

    def draw_balance_too_low(self):
        if self.bet_five or self.bet_ten or self.bet_twenty and self.bet_amount > self.wallet:
            low_balance_text = "No No No. Da. Da li imas pare? \n Balance too low for this bet. No"
            low_balance_surface = small_font.render(low_balance_text,True,(255,255,255))
            low_balance_rect = low_balance_surface.get_rect(center = (width/4,height/2))
            screen.blit(low_balance_surface,low_balance_rect)
        else:
            pass
```

```python
class MAIN:
    def __init__(self):
        self.deck = DECK()
        self.player = PLAYER()
        self.table = TABLE()
        self.menu = MENU()

    def draw_elements(self):
        self.player.draw_wallet()
        self.player.draw_bet_amount()
        self.player.draw_player_name()
        self.player.draw_dealer_name()
        self.table.draw_chips()
        self.table.draw_circle()
        self.menu.draw_menu_button()

    def draw_final_screen(self):
        self.player.draw_dealer_hand()
        self.player.draw_player_hand()
        self.player.draw_player_hand_value()
        self.player.draw_dealer_hand_value()
        self.table.fill_functions()
        self.table.draw_next_round()
        self.menu.draw_menu_button()

    def draw_game_over_screen(self):
        self.table.draw_game_over()
        self.table.draw_restart()
        self.menu.draw_menu_button()

    def update(self):
        self.player.place_bet()

    def deal_first_cards(self):
        if len(self.deck.shuffled_deck) < 10:
            self.deck.shuffled_deck = self.deck.live_deck()
        
        for i, card in enumerate(self.deck.shuffled_deck):
            if len(self.player.hand) == 0:
                self.player.hand.append(self.deck.shuffled_deck[0])
                self.deck.shuffled_deck.pop(0)
                self.player.dealer_hand.append(self.deck.shuffled_deck[0])
                self.deck.shuffled_deck.pop(0)
                self.player.hand.append(self.deck.shuffled_deck[0])
                self.deck.shuffled_deck.pop(0)
        self.player.calculate_player_hand_value()
        self.player.calculate_dealer_hand_value()

    def dealer_logic(self):
        global round_over
        while int(self.player.dealer_hand_value) < 17:
            card = self.deck.shuffled_deck.pop(0)
            pygame.time.delay(250)
            self.player.dealer_hand.append(card)
            self.player.dealer_hand_value = self.player.calculate_dealer_hand_value()
            self.player.draw_dealer_hand()
            self.player.draw_dealer_hand_value()
            print(self.player.dealer_hand)
            print(self.player.dealer_hand_value)
            pygame.display.update()
        
        if int(self.player.dealer_hand_value) > 21:
            self.player.draw_dealer_hand()
            self.player.draw_dealer_hand_value()
            pygame.display.update()
            self.player.dealer_bust = True
            round_over = True
        elif int(self.player.dealer_hand_value) == 21:
            self.player.draw_dealer_hand()
            self.player.draw_dealer_hand_value()
            pygame.display.update()
            self.player.dealer_blackjack = True
            self.player.dealer_stay = True
            round_over = True
        else:
            self.player.draw_dealer_hand()
            self.player.draw_dealer_hand_value()
            pygame.display.update()
            self.player.dealer_stay = True
            round_over = True

            
    def restart(self):
        global next_round
        global round_over
        global turn_over
        global hit
        global stay
        global double
        global split
        global dealer_turn
        global game_over
        global restart
        global same_cards
        self.player.bet_amount = int(0)
        self.player.wallet += 100
        self.player.hand.clear()
        self.player.dealer_hand.clear()
        self.player.bet_deducted = False
        self.player.bet_confirmed = False
        self.player.dealer_hit = False
        self.player.dealer_stay = False
        self.player.dealer_blackjack = False
        self.player.dealer_bust = False
        self.player.dealer_wins = False
        self.player.player_stay = False
        self.player.player_bust = False
        self.player.player_blackjack = False
        self.player.player_double = False
        self.player.player_wins = False
        self.player.push = False
        self.player.hand_calculation_done = False
        next_round = False
        round_over = False
        turn_over = False
        hit = False
        stay = False
        double = False
        split = False
        dealer_turn = False
        game_over = False
        restart = False
        same_cards = False

    def next_round(self):
        global next_round
        global round_over
        global turn_over
        global hit
        global stay
        global split
        global double
        global dealer_turn
        global same_cards
        self.player.bet_amount = int(0)
        self.player.hand.clear()
        self.player.dealer_hand.clear()
        self.player.bet_deducted = False
        self.player.bet_confirmed = False
        self.player.dealer_hit = False
        self.player.dealer_stay = False
        self.player.dealer_blackjack = False
        self.player.dealer_bust = False
        self.player.dealer_wins = False
        self.player.player_stay = False
        self.player.player_bust = False
        self.player.player_blackjack = False
        self.player.player_double = False
        self.player.player_wins = False
        self.player.push = False
        self.player.hand_calculation_done = False
        next_round = False
        round_over = False
        turn_over = False
        hit = False
        stay = False
        double = False
        split = False
        dealer_turn = False
        same_cards = False
        

    def player_bust(self):
        self.table.draw_bust()
        global round_over
        round_over = True
        self.player.player_bust = True

    def hit(self):
        self.player.hand.append(self.deck.shuffled_deck[0])
        self.deck.shuffled_deck.pop(0)
        self.player.calculate_player_hand_value()
        global hit
        hit = False

    def double(self):
        self.player.wallet -= self.player.bet_amount
        self.player.bet_amount += self.player.bet_amount
        self.player.player_double = False

    def split(self):
        new_hand_card = self.player.hand[1]
        self.player.hand_2.append(new_hand_card)
        self.player.hand.pop(1)

    def calculate_returns(self):
        if self.player.dealer_bust:
            if len(self.player.hand) == 2 and int(self.player.hand_value) == 21:
                self.player.wallet += (self.player.bet_amount * 2.5)
                self.player.player_blackjack = True
                self.player.player_wins = True
                self.player.hand_calculation_done = True
                pygame.time.delay(500)
            else:
                self.player.wallet += (self.player.bet_amount * 2)
                self.player.player_wins = True
                self.player.hand_calculation_done = True
        elif not self.player.dealer_bust:
            if int(self.player.hand_value) == int(self.player.dealer_hand_value):
                self.player.wallet += self.player.bet_amount
                self.player.push = True
                self.player.hand_calculation_done = True
            elif len(self.player.hand) == 2 and int(self.player.hand_value) == 21 and int(self.player.hand_value) > int(self.player.dealer_hand_value):
                self.player.wallet += (self.player.bet_amount * 2.5)
                self.player.player_blackjack = True
                self.player.player_wins = True
                self.player.hand_calculation_done = True
            elif len(self.player.hand) > 2 and int(self.player.hand_value) == 21 and int(self.player.hand_value) > int(self.player.dealer_hand_value):
                self.player.wallet += (self.player.bet_amount * 2)
                self.player.player_wins = True
                self.player.hand_calculation_done = True
            elif int(self.player.hand_value) < 21 and int(self.player.hand_value) > int(self.player.dealer_hand_value):
                self.player.wallet += (self.player.bet_amount * 2)
                self.player.player_wins = True
                self.player.hand_calculation_done = True
            elif int(self.player.hand_value) < 21 and int(self.player.hand_value) < int(self.player.dealer_hand_value):
                self.player.dealer_wins = True
                self.player.hand_calculation_done = True
        else:
            print("error in hand calculations")

    def calculate_game_over(self):
        if self.player.dealer_wins and int(self.player.wallet) == 0:
            global game_over
            game_over = True
        elif self.player.player_bust and int(self.player.wallet) == 0:
            game_over = True


        
```

```python
run = True
active_input_box = False
text = ""
current_background = welcome_background

next_round = False
round_over = False
turn_over = False
hit = False
stay = False
double = False
same_cards = False
split = False
dealer_turn = False
game_over = False
restart = False
menu = False

while run:
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            pygame.quit()
            sys.exit()
            run = False
        elif event.type == SCREEN_UPDATE:
            main_game.update()
        elif event.type == pygame.MOUSEBUTTONDOWN:
            if enter_name_rect.collidepoint(event.pos):
                active_input_box = True
            else:
                active_input_box = False
            if main_game.menu.menu_rect.collidepoint(event.pos):
                menu = True
            if len(main_game.player.hand) == 0:
                if main_game.menu.menu_close_rect.collidepoint(event.pos):
                    menu = False
                elif main_game.table.five_chip_rect.collidepoint(event.pos):
                    main_game.player.bet_five()
                    main_game.player.bet_confirmed = False
                elif main_game.table.ten_chip_rect.collidepoint(event.pos):
                    main_game.player.bet_ten()
                    main_game.player.bet_confirmed = False
                elif main_game.table.twenty_chip_rect.collidepoint(event.pos):
                    main_game.player.bet_twenty()
                    main_game.player.bet_confirmed = False
                elif main_game.table.bet_rect.collidepoint(event.pos) and not main_game.player.bet_confirmed:
                    main_game.player.place_bet()
                    main_game.player.bet_confirmed = True
                elif main_game.table.clear_rect.collidepoint(event.pos):
                    main_game.player.clear_bet()
            elif len(main_game.player.hand) >= 2:
                if main_game.menu.menu_close_rect.collidepoint(event.pos):
                    menu = False
                elif main_game.table.hit_rect.collidepoint(event.pos) and round_over == False:
                    hit = True
                elif main_game.table.stand_rect.collidepoint(event.pos) and round_over == False:
                    stay = True
                elif main_game.table.double_rect.collidepoint(event.pos) and len(main_game.player.hand) == 2 and round_over == False:
                    double = True
                elif main_game.table.next_round_rect.collidepoint(event.pos) and round_over == True and game_over == False:
                    next_round = True                
                elif main_game.table.restart_rect.collidepoint(event.pos) and round_over == True and game_over == True:
                    restart = True

        elif event.type == pygame.KEYDOWN:
            if active_input_box:
                if event.key == pygame.K_RETURN:
                    main_game.player.enter_name(text)
                    active_input_box = False
                    text = ""
                    current_background = main_background
                elif event.key == pygame.K_BACKSPACE:
                    text = text[:-1]
                else:
                    text += event.unicode
        
        elif menu == True:
            if main_game.menu.red_table_covering_rect.collidepoint(event.pos):
                current_background = main_background_red
            elif main_game.menu.blue_table_covering_rect.collidepoint(event.pos):
                current_background = main_background_blue
            elif main_game.menu.green_table_covering_rect.collidepoint(event.pos):
                current_background = main_background_green

    screen.fill([255,255,255])
    screen.blit(current_background,(0,0))
    if main_game.player.name is None:
        enter_name_text = "Enter your name"
        enter_name_surface = font.render(enter_name_text, True, (255,255,255))
        enter_name_rect = enter_name_surface.get_rect(topleft = (250,200))
        input_box_surface = pygame.Surface((250,50))
        input_box_surface.fill((255,255,255))  

        input_text_surface = font.render(text,True,(0,0,0))
        input_text_rect = input_text_surface.get_rect(center = input_box_surface.get_rect().center)
        input_box_rect = input_box_surface.get_rect(topleft=(250, enter_name_rect.bottom + 10))
        input_box_surface.blit(input_text_surface,input_text_rect)
        screen.blit(enter_name_surface,(250,200))               
        screen.blit(input_box_surface,(enter_name_rect.left,enter_name_rect.bottom))

        if input_box_rect.collidepoint(pygame.mouse.get_pos()):
            active_input_box = True
        else:
            active_input_box = False

    else:
        main_game.draw_elements()
        if menu == True:
            main_game.menu.draw_menu()
            print(pygame.mouse.get_pos())
        elif len(main_game.player.hand) == 0 and main_game.player.bet_deducted == False:
            place_your_bet_text = "Place your bet"
            place_your_bet_surface = large_font.render(place_your_bet_text, True, (255,255,255))
            place_your_bet_rect = place_your_bet_surface.get_rect(center=(width/2,height/2.5))
            screen.blit(place_your_bet_surface,(place_your_bet_rect.topleft,place_your_bet_rect.topleft))
            screen.blit(main_game.table.bet_surface, main_game.table.bet_rect)
            main_game.table.draw_clear()
            if main_game.player.bet_amount > 0:
                bet_image,bet_rect = main_game.player.draw_table_bet()
                screen.blit(bet_image,bet_rect)

        elif len(main_game.player.hand) == 0 and main_game.player.bet_deducted == True:
            main_game.deal_first_cards()
            main_game.player.draw_dealer_hand()
            main_game.player.draw_player_hand()
            main_game.player.draw_player_hand_value()
            main_game.player.draw_dealer_hand_value()
            main_game.table.draw_functions()

        elif len(main_game.player.hand) >= 2 and main_game.player.bet_deducted == True and next_round == False and round_over == False:
            main_game.player.draw_dealer_hand()
            main_game.player.draw_player_hand()
            main_game.player.draw_player_hand_value()
            main_game.player.draw_dealer_hand_value()
            main_game.table.draw_functions()

            if len(main_game.player.hand) == 2 and int(main_game.player.hand_value) == 21:
                dealer_turn = True

            if int(main_game.player.hand_value) <= 21 and not dealer_turn:
                if stay:
                    main_game.player.draw_dealer_hand()
                    main_game.player.draw_dealer_hand_value()
                    pygame.display.update()
                    dealer_turn = True
                        
                if hit:
                    main_game.hit()
                    main_game.player.draw_player_hand()
                    main_game.player.draw_player_hand_value()
                    pygame.display.update()
                if double:
                    main_game.double()
                    main_game.hit()
                    main_game.player.draw_player_hand()
                    dealer_turn = True
                    pygame.display.update()

            elif int(main_game.player.hand_value) > 21:
                main_game.player_bust()
                pass
            
            elif int(main_game.player.hand_value) <= 21 and dealer_turn and not main_game.player.dealer_stay and not main_game.player.dealer_bust:
                main_game.dealer_logic()

        #Here, we will calculate the returns & process payment
        elif len(main_game.player.hand) >= 2 and main_game.player.bet_deducted == True and next_round == False and round_over == True and not main_game.player.player_bust and not main_game.player.hand_calculation_done:
            main_game.calculate_returns()

        #Then we will print the final screens depending on the outcome of the game
        elif len(main_game.player.hand) >= 2 and main_game.player.bet_deducted == True and next_round == False and round_over == True and main_game.player.player_bust:
            main_game.calculate_game_over()
            if game_over and not restart:
                main_game.table.draw_game_over()
                main_game.table.draw_restart()
            elif game_over and restart:
                main_game.restart()              
            elif not game_over:
                main_game.draw_final_screen()
                main_game.table.draw_bust()
        elif len(main_game.player.hand) >= 2 and main_game.player.bet_deducted == True and next_round == False and round_over == True and main_game.player.dealer_bust:
            main_game.draw_final_screen()
            main_game.table.draw_winner()

        elif len(main_game.player.hand) >= 2 and main_game.player.bet_deducted == True and next_round == False and round_over == True and main_game.player.dealer_wins:
            main_game.calculate_game_over()
            if game_over and not restart:
                main_game.table.draw_game_over()
                main_game.table.draw_restart()
            elif game_over and restart:
                main_game.restart()
            elif not game_over:
                main_game.draw_final_screen()
                main_game.table.draw_dealer_wins()

        elif len(main_game.player.hand) >= 2 and main_game.player.bet_deducted == True and next_round == False and round_over == True and main_game.player.player_wins and not main_game.player.player_blackjack:
            main_game.draw_final_screen()
            main_game.table.draw_winner()

        elif len(main_game.player.hand) >= 2 and main_game.player.bet_deducted == True and next_round == False and round_over == True and main_game.player.player_wins and main_game.player.player_blackjack:
            main_game.draw_final_screen()
            main_game.table.draw_blackjack()

        elif len(main_game.player.hand) >= 2 and main_game.player.bet_deducted == True and next_round == False and round_over == True and main_game.player.push:
            main_game.draw_final_screen()
            main_game.table.draw_push()

        elif len(main_game.player.hand) >= 2 and main_game.player.bet_deducted == True and next_round == True and round_over == True:
            main_game.next_round()

    pygame.display.update()
    clock.tick(60)            
```
